## Artificial neural network example, classification, diabetes dataset

```python
# first neural network with keras tutorial
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
ds = pd.read_csv('datasets_228_482_diabetes_nh.csv')
X = ds.iloc[:, 0:8].values
y = ds.iloc[:,8 ].values
Xtr, Xtst, ytr, ytst = train_test_split(X, y, test_size=0.2 )
```

```python
print(Xtr)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(15, input_dim=8, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
print(type(model))
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(Xtr, ytr, epochs=200, batch_size=10)
# evaluate the keras model
_, accuracy = model.evaluate(Xtst, ytst, verbose=0)
print('Accuracy: %.2f ' % (accuracy*100))
```

```python
yprd = model.predict(Xtst)
print(yprd)   b
```

```python
yprd = np.round(yprd,0)
print(yprd,'\n\n\n')
print(confusion_matrix(ytst, yprd))
print('\n\n', classification_report(ytst, yprd))
```
